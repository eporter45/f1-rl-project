```python
import numpy as np
import pandas as pd
from pathlib import Path
import glob
import os
import shutil

# Set project root dynamically
project_root = Path().resolve().parent
tyre_dir = project_root / 'tyre_data'
data_dir = project_root/'data'
```

```python
# Get data file lists from data folder
file_list1 = glob.glob(str(tyre_dir / '*.csv'))
file_list2 = glob.glob(str(data_dir / '*.csv'))

# Load data into dataframes
dfs_tyre = {Path(file).stem: pd.read_csv(file) for file in file_list1}

# Create backup directory
backup_dir = project_root / 'tyre_data_backup'
os.makedirs(backup_dir, exist_ok=True)

# Copy original files to backup
for file in tyre_dir.glob('*.csv'):
    shutil.copy(file, backup_dir / file.name)

print(f"Backed up {len(list(backup_dir.glob('*.csv')))} files to {backup_dir}")

# Load data into dataframes
dfs_data = {Path(file).stem: pd.read_csv(file) for file in file_list2}
```

```python
race_dict = {
    '2015_1_australia': (926, 1),
    '2015_2_malaysia': (927, 2),
    '2015_3_china': (928, 17),
    '2015_4_bahrain': (929, 3),
    '2015_5_spain': (930, 4),
    '2015_6_monaco': (931, 6),
    '2015_7_canada': (932, 7),
    '2015_8_austria': (933, 70),
    '2015_9_britain': (934, 9),
    '2015_10_hungary': (936, 11),
    '2015_11_belgium': (937, 13),
    '2015_12_italy': (938, 14),
    '2015_13_singapore': (939, 15),
    '2015_14_japan': (940, 22),
    '2015_15_russia': (941, 71),
    '2015_16_usa': (942, 69),
    '2015_17_mexico': (943, 32),
    '2015_18_brazil': (944, 18),
    '2015_19_abudhabi': (945, 24),
    
    '2016_1_australia': (948, 1),
    '2016_2_bahrain': (949, 3),
    '2016_3_china': (950, 17)
}
```

```python
lap_times = dfs_data['lap_times']
# Step 1: Collect all valid driver IDs from lap times for the specified races
valid_driver_ids = set(
    lap_times.loc[
        lap_times['raceId'].isin([race[0] for race in race_dict.values()])
    ]['driverId'].unique()
)
valid_driver_ids = set(map(int, valid_driver_ids))
print(valid_driver_ids)

filtered_driver_map = {driver_id: name for driver_id, name in driver_map.items() if driver_id in valid_driver_ids}
print(f"Reduced driver map size: {len(filtered_driver_map)}")
print(filtered_driver_map)
driver_map = filtered_driver_map
```

```python
print(driver_map)
# First snippet — clean up before adding any integer values
driver_map = {v: k for k, v in driver_map.items()}

# Clean up driver_map to just last names and lowercase
cleaned_driver_map = {
    k: (v.split('_')[-1].lower()) if isinstance(v, str) else v
    for k, v in driver_map.items()
}

# Overwrite driver_map with cleaned names
driver_map = cleaned_driver_map
print(driver_map)
```

```python
driver_map['jnr'] = 832
print(driver_map)
driver_map = {v: k for k, v in driver_map.items()}
print(driver_map)
```

```python
for key in dfs_tyre.keys():
    if 'NAME' in dfs_tyre[key].columns:
        # Step 1: Split the name on whitespace and extract the last name
        last_name = dfs_tyre[key]['NAME'].str.split().str[-1]
        
        # Step 2: Convert the last name to lowercase and strip any extra spaces
        last_name = last_name.str.lower().str.strip()
        
        # Step 3: Assign it back to the DataFrame
        dfs_tyre[key]['lastName'] = last_name
        
        # Step 4: Map the last name to driverId
        dfs_tyre[key]['driverId'] = dfs_tyre[key]['lastName'].map(driver_map)

        # Debugging step — show mismatched names
        unmatched_names = last_name[~last_name.isin(driver_map.keys())]
        print(f"Unmatched names for key={key}: {unmatched_names.unique()}")
        
        # Debugging step — show mapped IDs
        print(f"Mapped IDs:\n{dfs_tyre[key][['lastName', 'driverId']].head(10)}")

        # Step 5: Handle unmapped drivers by filling with -1
        dfs_tyre[key]['driverId'] = dfs_tyre[key]['driverId'].fillna(-1).astype(int)
```

```python
import os

# Define output directory
output_dir = project_root / "cleaned_tyre_data"
os.makedirs(output_dir, exist_ok=True)

# Loop through race_dict to save using race_id as the filename
for key, (race_id, _) in race_dict.items():
    if key in dfs_tyre:
        output_file = output_dir / f"{race_id}.csv"
        dfs_tyre[key].to_csv(output_file, index=False)

print(f"✅ Cleaned tyre data saved to {output_dir} using race_id as filename")
```
